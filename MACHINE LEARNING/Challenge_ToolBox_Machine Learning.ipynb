**Todas las funciones deben llevar comentado el cdigo y deben incluir un docstring especificando su uso**, segn la siguiente gua:

Un "docstring" en Python es una cadena de texto utilizada para documentar mdulos, clases, mtodos y funciones. El propsito de un docstring es explicar brevemente qu hace la funcin y cmo se debe utilizar. Los docstrings se encierran entre tres comillas dobles (""") al inicio y al final, y se colocan justo despus de la definicin de la funcin. Aqu hay un formato tpico de un docstring para una funcin en Python:

```python
def mi_funcion(param1, param2):
    """
    Descripcin breve de lo que hace la funcin.

    Argumentos:
    param1 (tipo): Descripcin de param1.
    param2 (tipo): Descripcin de param2.

    Retorna:
    tipo: Descripcin de lo que retorna la funcin.
    """
    
    # El cuerpo de la funcin va aqu
    pass
```

En este formato:

1. **Primera lnea**: Breve descripcin de la funcin.
2. **Seccin de Argumentos**: Lista cada parmetro, su tipo esperado y una breve descripcin de su propsito.
3. **Seccin de Retorna**: Describe el tipo y el propsito del valor que la funcin retorna.
